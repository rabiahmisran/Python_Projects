```python
# Print The Largest

a = 19
b = 45
c = 27
d = 141
e = 67

largest = 0

if a>largest:           # 19>0 (t)
  largest = a           # Largest = 19
if b>largest:           # 45>19 (t)
  largest = b           # Largest = 45
if c>largest:           # 27>45 (f)
  largest = c           # Largest = 45
if d>largest:           # 141>45 (t)
    largest = d         # Largest = 141
if e>largest:           # 67>141 (f)
  largest = e           # Largest = 141

print(f'Largest is:{largest}')
```

```python
# One Liner if else
# ternary operator

# if_condition_is_true  if  condition   else  value_if_condition_is_false

num = int(input('Enter Any Number  : '))
result = 'Even Number'  if  num%2==0 else 'Odd Number'
print(result)
```

```python
# Chained Comparison
# help us to define a range
# Lower Limit <= value  <=  Upper Limit

print("Grading Allotment Application")
marks = int(input("Enter Your Marks :"))   # 83

# here 85 <= marks <= 100 this expression means
#  a.) the value is less than or equal to 100    c1     is 93 <=100 (t)
#  b.) 85 means that I am less the define value  c2     is 85 is less than 93 (t)

if 85 <= marks <= 100:       #  83<=100 (t)   85 is it less than 83 (f)
    print('Grade A')
elif 75 <= marks <= 84:      # 83<=84  (t)    75 is it less than 83 (t)
    print('Grade B')
```

```python
print('Hi! this is Your BMI Calculator')

name = input('Enter Your Name:')
weight = float(input('Enter your weight in kg :  '))
height = float(input('Enter your height in cm :  '))
meter = height/100
BMI = weight/(meter*meter)
print(f'Your BMI is {BMI:.2f}')
if BMI<18.5:
  print('Your BMI is Underweight😞')
  print('Eat More')
elif 18.5<= BMI <=24.9:
  print('Normal😍')
  print('Keep it up')
elif 25<= BMI <=29.9:
  print('Overweight😲')
  print('Eat Healthy')
else:
  print('You are Obese😓')
  print('Please Eat Healthy')
```
