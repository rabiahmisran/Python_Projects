```python
# range()
# it's a built-in function and it is used to generate
# a sequence of Integer Numbers

# a.) range(x)
# it is used to generate a sequence of Integer Numbers
# starting from zero till x-1

print(list(range(10)))

# b.) range(x,y)
# it is used to generate a sequence of Integer Numbers
# starting from zero x till y-1

print(list(range(1,20)))

# c.) range(x,y,stepCount/Interval)
# it is used to generate a sequence of Integer Numbers
# starting from x till y-1 based on stepCount

print(list(range(1,11,3)))

# print all the even numbers till 100 using range ??

print(list(range(0,101,2)))
```

```python
# while loop

# used to repeat a block of code several times

# until a certain condition is met

# syntax

# initialization           [optional]

# while condition/expression:

#   code to be executed

#   increment / decrement  [optional]

# WHILE WILL CHECK THE CONDITION FIRST

# IF THE CONDITION IS TRUE , WHILE WILL EXECUTE ITS BODY

# AFTER WHILE FINISH EXECUTING THE BODY , CONDITION WILL BE CHECKED AGAIN

# IF THE CONDITION IS TRUE , WHILE WILL AGAIN EXECUTE ITS BODY

i = 1                  # 1ST ITER  i = 1    2ND ITER  i = 2     3RD ITER  i = 3   4TH ITER
While i<=3:            # 1<=3 (t)            2<=3 (t)            3<=3 (t)         4<=3 (FALSE)
    print(i)           # 1                   2                   3
    i = i + 1          # i = 1 + 1 ==> 2     i = 2 + 1 ==> 3     i = 3 + 1 ==> 4
```

```python
# PRINT MATHEMATICAL TABLE OF A NUMBER 13
# USING WHILE LOOP

num = 13
i = 1
while i<=14:
  print(f'{num}x{i}={num*i}')
  i = i + 1
```

```python
# Use while loop when the number of iterations are not known

num = 121
temp = num
sum = 0

while num!=0:           # 1st Iteration  121!=0(t)   2nd Iteration  12!=0(t)   3rd Iteration 1!=0(t)   4th Iteration 0!=0 (f)
  rem = num % 10        #  rem = 1                    rem = 2                    rem = 1
  sum = sum*10 + rem    #  sum = 1                    sum = 12                   sum = 121
  num = num // 10       #  num = 12                   num = 1                    num = 0

if temp == sum:
  print('Palindrome Number')
else:
  print('Not A Palindrome')
```

```python
num = 12345
temp = num
sum = 0

while num>0:           # 1st Iteration  12345>0(t)   2nd Iteration  1234>0(t)   3rd Iteration 123>0(t)   4th Iteration 12>0 (f)   5th Iteration 1>0 (f)
  rem = num % 10        #  rem = 5                     rem = 4                     rem = 3                    rem = 2              rem = 1   print(1%10)
  sum = sum*10 + rem    #  sum = 5                     sum = 54                    sum = 543                  sum = 5432           sum = 54321
  num = num // 10       #  num = 1234                  num = 123                   num = 12                   num = 1              num = 0    print(1//10)

if temp == sum:   # 12345 ==54321
  print('Palindrome Number')
else:
  print('Not A Palindrome')
```

```python
# break and continue statement

# it is used to skip the current execution of a LOOP

# break and continue should be defined inside decision making

for i in range(1,11):
  if i == 5 or i == 9:
    continue
  print(i)
```

```python
# Guess a number game
# Objective : our machine will generate one random number
# we have to guess the number in 4 attempts

import random
guessAttempt = 1

print('🐰 Guess A Number Game')
print('Enter Your Name')
name = input()
print(f'Welcome User {name}🤩')
print('I am thinking of a number between 1 and 13 🧐. Please Wait...')
randomNum = random.randint(1,13)
print('Number Is Generated ✅ Guess The Number In 4 Attempts!')
while guessAttempt <= 4:
  print(f'Attempt{guessAttempt} Take A Guess.')
  guess = int(input())

  if guess == randomNum:
    print('🥳 Congratulations. Correct Guess💥')
    print(f'You have guess the number in {guessAttempt} attempt.')
    break

    if guess > randomNum:
      print('Your Guess is too High💨 Think of a Smaller Number')

      if guess < randomNum:
        print('Your Guess is too Low😒 Think of a Bigger Number')

  guessAttempt = guessAttempt + 1

  if guess!=randomNum:
    print('🚫 GAME OVER 🚫')
    print(f'The Correct Number was {randomNum}')
```
