```python
# Exception handling in Python is a mechanism that allows you to handle runtime errors,
# or exceptions, in a controlled way, rather than letting your program crash unexpectedly.
# Python provides a way to catch errors, respond to them, and allow your program to continue
# running smoothly even if something goes wrong.
# try except finally
# try : keep that block of code inside try , which might give you an error
# except : catch that error
```

```python
try:
    val1 = int(input('Enter First Value :'))
    val2 = int(input('Enter Second Value :'))
    res = val1/val2
    print(f'After Division Result Will Be {res:.1f}')
except:
    print('Something Goes Wrong...')
    print('Make Sure to Provide Only Integer Input or Second Value Can\'t Be Zero')
print('Rest of The Code')
```

```python
# Using An Exception Class
# if you need to know the exact reason for an error use Exception class

try:
    val1 = int(input('Enter First Value :'))
    val2 = int(input('Enter Second Value :'))
    res = val1/val2
    print(f'After Division Result Will Be {res:.1f}')
except Exception as e:
    print('Something Goes Wrong...')
    print('Error :', e)
print('Rest of The Code')
```

```python
# Using dedicated Exception Class

try:
    val1 = int(input('Enter First Value :'))
    val2 = int(input('Enter Second Value :'))
    res = val1/val2
    print(f'After Division Result Will Be {res:.1f}')
except (ZeroDivisionError, ValueError) as e:
    print('Something Goes Wrong...')
    print('Error :', e)
print('Rest of The Code')
```

```python
# finally
# finally execute always

# Case No. 1 : WHEN IS NO EXCEPTION

try:
    val1 = int(input('Enter First Value :'))
    val2 = int(input('Enter Second Value :'))
    res = val1/val2
    print(f'After Division Result Will Be {res:.1f}')
except Exception as e:
    print('Something Goes Wrong...')
    print('Error :', e)
finally:
    print('I WILL EXECUTE ALWAYS 😎')
print('Rest of The Code')
```

```python
# Case No. 3 : WHEN THERE IS AN EXCEPTION BUT NOT HANDLED PROPERLY

try:
    val1 = int(input('Enter First Value :'))
    val2 = int(input('Enter Second Value :'))
    res = val1/val2
    print(f'After Division Result Will Be {res:.1f}')
except ImportError as e:
    print('Something Goes Wrong...')
    print('Error :', e)
finally:
    print('I WILL EXECUTE ALWAYS 😎')
print('Rest of The Code')
```
