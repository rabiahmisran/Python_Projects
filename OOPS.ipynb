```python
# OOPS
# OOP stands for Object-Oriented Programming.
# Object-oriented programming (OOP) is a programming paradigm based on the
# concept of "objects"
# Procedural programming is about writing procedures or methods that perform operations
# on the data, while object-oriented programming is about creating objects
# that contain both data and methods.
# Object-oriented programming has several advantages over procedural programming:
# a.) OOP is faster and easier to execute
# b.) OOP provides a clear structure for the programs
# c.) OOP helps to keep the Java/Python code DRY "Don't Repeat Yourself", and makes the code easier
#     to maintain, modify and debug
# d.) OOP makes it possible to create full reusable applications with less code and
#     shorter development time
# Class
# A class is a blueprint that defines the variables and the methods common to all objects.
# a class is a collection of methods and variables

# Object
# An object is called an instance of a class.

# Self Keyword
# The self parameter is a reference to the current instance of the class, and is used to
# access variables that belongs to the class.
# The first parameter of each method in a class must be the self ,
# which refers to the calling object. However, you can give any name to the first
# parameter, not necessarily self.
# __init__ in Python
#  Constructors are used to initializing the object’s state. The task of constructors is
# to initialize(assign values) to the data members of the class when
# an object of the class is created.
# The __init__() function is called automatically every time the class is being used to create
# a new object.
```

```python
# class in Python
# a class is a collection of methods/functions and variables
# in python a class can be defined by using a class keyword

class myClass:

    a = 10         # member variable
    b = 20

    def func1():   # member function
        print('I am the method of a class')

    def func2():
        print('I am the another of a class')

print(myClass.a)  # . dot is a member access operator
myClass.func1()
```

```python
class myClass:

    a = 10         # member variable
    b = 20

    def func1():   # member function
        print('I am the method of a class')

    def func2():
        print('I am the another of a class')

# In OOPs in order to access class variables and functions
# we need to define the Object of a class

# syntax for defining the object
# objName = className()

# object can access the class variables and functions
# in python a class can have n number of objects

obj1 = myClass()    # creating an object of a class
obj2 = myClass()    # creating an another object of a class

print(obj1.a)
print(obj1.b)

print(obj2.a)
print(obj2.b)
```

```python
# SELF KEYWORD 
# SELF REPRESENT THE CURRENT CALLING OBJECT

# =====IMPORTANT RULE======
# IN PYTHON IF YOU ARE DEFINING A FUNCTION INSIDE THE CLASS
# THE FIRST PARAMETER OF THAT FUNCTION NEEDS TO BE SELF

class myClass:        # 3.) giving your class a clear idea that obj1 wants to use this func1() now

    a = 10  
    b = 20

    def func1(self):     # 2.) this self value will be replaced by obj1    func1(obj1)    func1(obj2)
        print('I am the method of a class')

    def func2(self):
        print('I am the another method of a class')

# WHY SELF NEED TO BE WRITTEN INSIDE THE CLASS FUNCTION
# REASON 1 : A CLASS CAN HAVE N NUMBER OF OBJECTS


obj1 = myClass()  
obj2 = myClass()

obj1.func1()    # 1.) here obj1 is calling the func1()
obj2.func1()    # here obj2 is calling the func1()
```

```python
# Design an application for me using Procedural Programming
# User Registeration System
 
# Task1
# Take 3 information from 3 users
# user1      user2    user3
# name       name     name
# email      email    email
# mobno      mobno    mobno
 
# Task2
# Print all this information
 
# Task3
# Print only the second user information
 
# Task4
# Print the email and mobno of 3rd User

users = []
 
for i in range(1,4):
    print("Enter details for User", i)
 
    name = input("Enter Name: ")
    email = input("Enter Email: ")
    mobno = input("Enter Mobile Number: ")
 
    user = [name, email, mobno]
    users.append(user)
 
    print()
 
# Task 2: Print all information
print("All Users Information:")
for user in users:
    print("Name:", user[0], "Email:", user[1], "Mobile:", user[2])
 
print()
 
# Task 3: Print only second user information
print("Second User Information:")
print("Name:", users[1][0])
print("Email:", users[1][1])
print("Mobile:", users[1][2])
 
print()
 
# Task 4: Print email and mobile number of third user
print("Third User Email and Mobile:")
print("Email:", users[2][1])
print("Mobile:", users[2][2])
```

```python
# Python OOP provides us a special function named as __init__()
#  __init__() can help you to provide diff diff data value to a class variables
#  this special function will call itself the moment when you create the
#  object of a class
```

```python
class userInfo:
 
    def __init__(self,n,e,mn):    # us3  , n = Smith  ,   e = sm@gmail.com  , mn = 60690231

        self.name = n

        self.email = e

        self.mobno = mn
 
    def printDetails(self):                            # printDetails(us1)                       printDetails(us2)      

        print(self.name,self.email,self.mobno)         #   us1.name us1.email us1.mobno           us2.name us2.email us2.mobno

us1 = userInfo("Auni","auni@gmail.com",60148411450)  # __init__ calls itself and this information ("John","j@gmail.com",60987654) will be passed to it automatically

us2 = userInfo("Aqil","aqil@gmail.com",601496957888) # __init__ calls itself and this information ("Alex","al@gmail.com",60453213) will be passed to it automatically

us3 = userInfo("Azmir","azmir@gmail.com",60149692740) # __init__ calls itself and this information ("Smith","sm@gmail.com",60690231) will be passed to it automatically
 
print("====Printing All Information====")

us1.printDetails()

us2.printDetails()

us3.printDetails()
 
print("\n====Printing Second User Information====")

us2.printDetails()
 
print("\n====Printing Email and MobNo Of 3rd User====")

print(us3.email)

print(us3.mobno)
 
```
